```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.cluster import AffinityPropagation
aff = AffinityPropagation(random_state = 5)
y_aff = aff.fit_predict(X)
```

```python
supervised = pd.DataFrame(dataset)
```

```python
supervised['Cluster_group'] = y_aff
```

```python
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data = supervised, x = supervised.columns[3], y = supervised.columns[4],
                   hue = supervised.columns[5], fit_reg = False, legend = True, legend_out = True)
```
